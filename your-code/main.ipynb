```python
repos_url = 'https://api.github.com/users/soniameller/repos'
response_repos = requests.get(repos_url,auth = HTTPBasicAuth('soniameller', 'f8795257cda18efbbce6e8a7d70166be9ab5dbe1'))
```

```python
response_repos.json()
```

```python
lang_endpoints = [repo['languages_url'] for repo in response_repos.json()]

lang_endpoints
```

```python
languages = [requests.get(lang_url,auth = HTTPBasicAuth('soniameller', 'f8795257cda18efbbce6e8a7d70166be9ab5dbe1')).json() for lang_url in lang_endpoints]

```

```python

lang_keys = [list(language.keys()) for language in languages ]

unique_languages =list(set([lang for lang_sublist in lang_keys for lang in lang_sublist]))

unique_languages
```
